# Stack

In Java, a Stack is a linear data structure that follows the **Last-In, First-Out** (LIFO) principle and is defined in the `java.util package`. Internally, it extends the `Vector` class.




```java
Stack<Integer> s = new Stack<>();
```

## Operation on Stack

* `push()` - as adding and return that new item
* `pop()` - as removing and return that new item
* `peek()` - looks at the top object (non-removal)
* `empty()` - test if the stack is empty
* `search()` - the 1-based position from the top of the stack where the object is located; the return value -1 indicates that the object is not on the stack.ak.



```java
// Push objects
s.push(5);
s.push(6);
s.push(7);
s.push(8);
s
```

```java
// Pop the top
s.pop();
```

```java
// Peek for the top
s.peek();
```

```java
s.empty();
```

```java
s.search(8);
```

```java
s.size();
```
