# Stack & Queue trong Java

## Stack

Trong Java, Stack (ngăn xếp) là một cấu trúc theo dạng LIFO (cái nào đến sau thì ra trước)



Một số hàm cơ bản có sẵn trong Java để làm việc với stack (xem thêm ở [Java Doc](https://docs.oracle.com/en/java/javase/11/docs/api/java.base/java/util/Stack.html))

* `empty()` (boolean) - Kiểm tra xem một stack có rỗng hoặc không
* `peek()` (void) - Lấy giá trị của object cuối cùng được thêm vào
* `pop()` (void) - Xóa đối tượng ở đầu ngăn xếp và trả về đối tượng đó dưới dạng giá trị của hàm.
* `push()` (void) - Đẩy một mục lên đầu ngăn xếp.
* `search(Object o)` (int) - Trả về vị trí bắt đầu từ 1 nơi có một đối tượng trên ngăn xếp này.


```java
// Khởi tạo một Stack
Stack<String> sampleStack = new Stack<String>();
```

```java
// Thêm phần tử vào Stack
sampleStack.push("A");
sampleStack.push("B");
sampleStack.push("C");
```

```java
// Hiển thị Stack ban đầu
System.out.println("Initial Stack: " + sampleStack); // Output: [A, B, C]
```

```java
// Xóa phần tử khỏi Stack
System.out.println("Removed Element: " + sampleStack.pop()); // Output: C
System.out.println("Stack after removal: " + sampleStack); // Output: [A, B]
```

## Queue

Trong Java, Queue (hàng đợi) là một cấu trúc dạng FIFO (cái nào đến trước thì ra trước)



Một số hàm cơ bản làm việc với Queue. (do Queue là một Interface nên khi khởi tạo các đối tượng cần phải sử dụng các "Known Implementing Classes", xem thêm ở [Java Doc](https://docs.oracle.com/javase/8/docs/api/java/util/Queue.html))

* `add()` (boolean) - Chèn phần tử đã chỉ định vào hàng đợi nếu có thể thực hiện ngay lập tức mà không vi phạm giới hạn dung lượng, trả về `true` nếu thành công và ném `IllegalStateException` nếu hiện không có không gian khả dụng.
* `element()` (Object) - Lấy giá trị nhưng không xóa phần đầu của hàng đợi.
* `offer(Object)` (boolean) - Chèn phần tử được chỉ định vào hàng đợi nếu có thể thực hiện ngay lập tức mà không vi phạm giới hạn dung lượng.
* `peek()` (Object) - Truy xuất nhưng không xóa phần tử đầu của hàng đợi này hoặc trả về `null` nếu hàng đợi trống.
* `pool()` (Object) - Truy xuất và xóa phần tử đầu của hàng đợi này hoặc trả về `null` nếu hàng đợi này trống.*
* `remove()` (Object) - Lấy và xóa phần tử đầu của hàng đợi. Có ném exception `NoSuchElementException` nếu hàng đợi đang trống.


```java
// Khởi tạo một Queue sử dụng ArrayDeque
ArrayDeque<String> sampleQueue = new ArrayDeque<String>();
```

```java
// Thêm phần tử vào Queue
sampleQueue.add("A");
sampleQueue.add("B");
sampleQueue.add("C");
```

```java
// Hiển thị Queue ban đầu
System.out.println("Initial Queue: " + sampleQueue); // Output: [A, B, C]
```

```java
// Xóa phần tử khỏi Queue
System.out.println("Removed Element: " + sampleQueue.remove()); // Output: A
System.out.println("Queue after removal: " + sampleQueue); // Output: [B, C]
```
