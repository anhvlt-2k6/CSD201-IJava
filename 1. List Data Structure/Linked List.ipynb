# Linked List

A linked structure is a collection of nodes storing data and links to other nodes

A linked list is a linear data structure composed of nodes, each node holding some information and a reference to another node in the list

Types of linked lists:

* Singly Linked list
* Doubly Linked listst



```java
LinkedList<String> cars = new LinkedList<>();
```

## Linked list methods

For many cases, the `ArrayList` is more efficient as it is common to need access to random elements in the list, but the `LinkedList` provides several methods to do certain operations more efficiently (alongside with `List` ADT):

* `addFirst()`
* `addLast()`
* `removeFirst()`
* `removeLast()`
* `getFirst()`
* `getLast()`


```java
cars.addFirst("Toyota");
cars.addFirst("Volvo");
```

## Singly Linked List implementation

A singly linked list is a list whose node includes two data fields: `info` and `next`. The info field is used to store information, and this is important to the user. The next field is used to link to its successor in this sequence




```java
class Node {
    int data;
    Node next;

    Node(int new_data) {
        this.data = new_data;
        this.next = null;
    }
}
```

```java
Node head = new Node(10);

// Link the second node
head.next = new Node(20);

// Link the third node
head.next.next = new Node(30);

// Link the fourth node
head.next.next.next = new Node(40);

// printing linked list
while (head != null) {
    System.out.print(head.data + " ");
    head = head.next;
}
```

## Doubly Linked List

In a doubly linked list, each node has two reference fields, one to the successor and one to the predecessor





```java
class Node {
    int data;
    Node prev, next;

    public Node(int data) {
        this.data = data;
        this.prev = null;
        this.next = null;
    }
}
```

```java
Node head = new Node(10);
Node next = new Node(20);

head.next = next;
head.prev = null;

next.prev = head;
next.next = null;

Node anotherNext = new Node(30);

next.next = anotherNext;

anotherNext.prev = next;
anotherNext.next = null;
```

```java
Node node = head;

while (node != null) {
    System.out.println("Current: " + node.data);
    node = node.next;
}
```
