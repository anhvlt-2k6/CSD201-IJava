# Tree

In computer science, a tree is an abstract model of a hierarchical structure

## Breadth-first Traversal of a tree

* **Breadth-first traversal** (Level-order traversal) is visiting each node starting from the lowest (or highest) level and moving down (or up) level by level, visiting nodes on each level from left to right (or from right to left).* **Depth-first traversal** starts at the root node and examines each branch as far as possible before backtracking




### Implement Tree as class per node

```java
class Node {
    int info;
    Node left, right;

    public Node(int info) {
        this.info = info;
        this.left = null;
        this.right = null;
    }
}
```

```java
Node root = new Node(10);
root.left = new Node(20);
root.right = new Node(30);
```
